```python
import pandas as pd
df = pd.read_csv('airport_reviews[1].csv')
df.head()
```

```python
!pip install markovify
import markovify

```

```python
#join individual reviews into 1 large text string
from itertools import chain
N=100
review_subset = df["content"][0:N]
text = "".join(chain.from_iterable(review_subset))
markov_chain_model = markovify.Text(text)
```

```python
#Generate 5 sentences using the Markov Chain model
for i in range(5):
  print(markov_chain_model.make_sentence())
```

```python
#generate 3 sentences with length no more than 140 characters
for i in range(3):
  print(markov_chain_model.make_short_sentence(140))
```
